```python
from sklearn.neighbors import LocalOutlierFactor
```

```python
from sklearn.datasets import make_moons
```

```python
X,y=make_moons(n_samples=300,noise=0.1)
```

```python
import numpy,pandas,matplotlib.pyplot 
```

```python
X_out=numpy.random.uniform(low=-3,high=3,size=(20,2))
X=numpy.vstack((X,X_out))
```

```python
df=LocalOutlierFactor(n_neighbors=20,contamination=0.1)
y_pred=df.fit_predict(X)
```

```python
X_inliners=X[y_pred==1]
X_outliers=X[y_pred==-1]
```

```python
# Separate normal data points>>inliers and outliers 
X_inliers = X[y_pred == 1]
X_outliers = X[y_pred == -1]

# Plotting
matplotlib.pyplot.title("Local Outlier Factor (LOF) on Moons Data")
matplotlib.pyplot.scatter(X_inliers[:, 0], X_inliers[:, 1], color='b', s=20, edgecolor='k', label='Normal data points')
matplotlib.pyplot.scatter(X_outliers[:, 0], X_outliers[:, 1], color='r', s=20, edgecolor='k', label='Outliers')
matplotlib.pyplot.legend()
matplotlib.pyplot.show()
```
