```python
import numpy as np
from sklearn.datasets import make_circles
```

```python
X,y=make_circles(n_samples=750,factor=0.3,noise=0.7)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
from sklearn.cluster import DBSCAN
```

```python
df=DBSCAN(eps=0.1)
```

```python
pred=df.fit_predict(X)
```

```python
pred
```

```python
plt.scatter(X[:,0],X[:,1],c=df.labels_)
```
