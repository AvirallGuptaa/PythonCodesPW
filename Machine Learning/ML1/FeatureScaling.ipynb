```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
```

```python
df
```

# STANDARDIZATION

```python
mean=np.mean(df["total_bill"])
std=np.std(df["total_bill"])
```

```python
std_data=[]
for i in list(df["total_bill"]):
    zscore=(i-mean)/std
    std_data.append(zscore)
```

```python
std_data
```

STANDARDIZTION USING LIB

```python
from sklearn.preprocessing import StandardScaler 
scaler=StandardScaler()
```

```python
pd.DataFrame(scaler.fit_transform(df[["total_bill","tip"]]),columns=["total_bill","tip"])
```

# NORMALIZATION

```python
from sklearn.preprocessing import MinMaxScaler
minmax=MinMaxScaler()
pd.DataFrame(minmax.fit_transform(df[["total_bill","tip"]]),columns=["total_bill","tip"])

```
