```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
```

```python
data=load_iris()
```

```python
print(data.DESCR)
```

```python
df=pd.DataFrame(data.data,columns=data.feature_names)
df["target"]=data.target
```

```python
df.head(5)
```

```python
df
```

```python
df.target.unique()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df_final=df[df["target"]!=2]
```

```python
df_final.target.unique()
```

```python
df_final
```

```python
X=df_final.iloc[:,:-1]
y=df_final.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=1)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
y_pred_probab=model.predict_proba(X_test)
```

```python
y_pred_probab
```

```python
model
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import roc_curve, auc
```

```python
fpr,tpr,threshold=roc_curve(y_test,y_pred_probab[:,1])
```

```python
fpr,tpr,threshold
```

```python
roc_auc=auc(fpr,tpr)
```

```python
roc_auc
```

```python
plt.figure(figsize=(8,6))
plt.plot(fpr,tpr,color="orange",label="ROC curve (area: %0.2f)"% roc_auc)
plt.plot([0,1],[0,1],color="blue",linestyle="--")
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.05])
```

```python
from sklearn.model_selection import KFold
cv=KFold(n_splits=5)
from sklearn.model_selection import cross_val_score
score=cross_val_score(model,X_train,y_train,cv=cv,scoring="accuracy" )
```
